## Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def Find_range(list_numbers):
  minimum_value = list_numbers[0] # setting the first value of the list as min value
  maximum_value = list_numbers[0] # setting the first value of the list as max value

  for num in list_numbers:
  # looping through the entire list to check for maximum_value and minimum_value
    if(num >= maximum_value):
      maximum_value = num
    elif(num <= minimum_value):
      minimum_value = num
  return(minimum_value, maximum_value)

mylist = [88,91,16,64,38,99,47,61]
range = Find_range(mylist)
print("Range:",range)
```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean_value(numbers):
  total = 0
  for i in numbers:
    total += i
  length = len(numbers)
  return total/length

num = [1,2,3,4,5,6]
result = mean_value(num)
print("Mean: ",result)
```

Q3. Write a function that returns the median value of a list of numbers

```python
def median_value(list_numbers):
  list_numbers = sorted(list_numbers)
  length = len(list_numbers)
  mid = length // 2
  if length % 2 == 1:
    return list_numbers[mid]  # odd length
  else:
    return (list_numbers[mid-1] + list_numbers[mid]) / 2    # even length

nums = [1,2,3,4]
print("Median: ",median_value(nums))
```

Q8. Write a function to return standard deviation of a list of numbers

```python
def standard_deviation(numbers):
  n = len(numbers)
  mean = sum(numbers) / n
  squared_diff = [(x - mean) ** 2 for x in numbers]
  variance = sum(squared_diff) / n
  return variance ** 0.5
x = [2,5,7,3,5,7,8,1]
print(standard_deviation(x))
```
