```python
import numpy as np
import pandas as pd
import sqlite3
import requests
```

```python
# Load launch data from SpaceX API
url='https://api.spacexdata.com/v4/launches'

response=requests.get(url)
data=response.json()
df_launch=pd.DataFrame(data)
df_launch
```

```python
df_launch.info()
```

###Step 1: Load SpaceX Launch Data from API

```python
df_launch.head()
```

```python
launch_df = pd.read_json(url)
launch_df.head()
```

```python
# Extract required columns
launch_df = launch_df[["name","date_utc","success","details","rocket"]]
launch_df.head()
```

```python
# Convert date_utc to datetime
launch_df["date_utc"] = pd.to_datetime(launch_df["date_utc"])
```

```python
# Extract year from date_utc
launch_df["year"] = launch_df["date_utc"].dt.year
```

```python
# Show cleaned data
launch_df.head()
```

###Step 2: Load Rocket Metadata

```python
ur2='https://api.spacexdata.com/v4/rockets'
rocket_df = pd.read_json(ur2)
rocket_df.head()
```

```python
# Extract required columns
rocket_df = rocket_df[["id", "name", "type", "active", "stages"]]
rocket_df.head()
```

###Step 3: Merge Launch and Rocket Data

```python
merged_df = pd.DataFrame(launch_df.merge(rocket_df, left_on="rocket", right_on="id"))
merged_df.head()
```

###Step 4: Add Simulated Country Information

```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_df["country"] = np.random.choice(countries, size=len(merged_df))
merged_df.head()
```

###Step 5: Store Merged Data in SQLite3

```python
# Create SQLite connection
conn = sqlite3.connect("spacex_launches.db")

# Save full merged DataFrame with ALL columns (including country)
merged_df.to_sql("launches", conn, if_exists="replace", index=False)
conn.commit()
merged_df
```

###Step 6: Run SQL Queries on the Data to analyze

```python
# Launches by Country
q1 = """SELECT country, COUNT(*) AS total_launches FROM launches
GROUP BY country ORDER BY total_launches DESC"""

pd.read_sql_query(q1, conn)
```

```python
# Which year had the highest number of launches
q2 = """SELECT year, COUNT(*) AS total_launches FROM launches
GROUP BY year ORDER BY total_launches DESC LIMIT 1"""

pd.read_sql_query(q2, conn)
```

```python
# Top 5 Missions by Launch Count
q3 = """SELECT name_x AS mission_name, COUNT(*) AS launch_count FROM launches
GROUP BY name_x ORDER BY launch_count DESC LIMIT 5"""

Launch_Count = pd.read_sql_query(q3, conn)
Launch_Count
```

```python
# Top Rocket
q4 = """SELECT name_y AS rocket_name, COUNT(*) AS rocket_count FROM launches
GROUP BY name_y ORDER BY rocket_count DESC LIMIT 1"""

Rocket_Count = pd.read_sql_query(q4, conn)
Rocket_Count
```
