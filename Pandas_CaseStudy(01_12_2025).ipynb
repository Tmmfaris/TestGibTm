###4) Check and remove duplicates

```python
# Check for duplicates
print("Number of Duplicate Rows:", df.duplicated().sum())

# Remove duplicate rows and update the DataFrame in-place
df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", df.shape)
```

###5) Rename columns ("Title" > "Movie_Title")

```python
df = df.rename(columns={"Title":"Movie_Title"})
df
```

6) Get unique genres

```python
df['Genre'].unique()
```

```python
unique_genres = df["Genre"].str.split(",").explode().unique()
unique_genres
```

###7) Extract number of genres per movie

```python
df["Num_Genres"] = df["Genre"].str.split(",").str.len()
df[["Genre", "Num_Genres"]]
```

###8) Handle missing values in 'Revenue (Millions)' and 'Metascore

```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].mean())
df[['Revenue (Millions)', 'Metascore']].isna().sum()
```

###9) Top 5 movies by Rating

```python
top5_movies = df.sort_values(by="Rating", ascending=False).head(5)
top5_movies[["Movie_Title", "Rating",]]
```

###10) Movies with highest/lowest revenue

```python
# Highest Revenue
highest_rev = df.loc[df["Revenue (Millions)"].idxmax()]
highest_rev[["Movie_Title", "Revenue (Millions)"]]
```

```python
# Lowest Revenue
lowest_rev = df.loc[df["Revenue (Millions)"].idxmin()]
lowest_rev[["Movie_Title", "Revenue (Millions)"]]
```

###16) Create a new column Profitability:

     a) Revenue divided by Duration

     b) Find top 10 most profitable movies


```python
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
df[["Movie_Title","Revenue (Millions)","Runtime (Minutes)","Profitability"]]
```

```python
# Find top 10 most profitable movies
top_10=df.nlargest(10,"Profitability")
top_10[["Movie_Title","Revenue (Millions)","Runtime (Minutes)","Profitability"]]
```

###17) Create a column that categorizes movies by rating:

      a) Rating < 6: Poor

      b) 6 â‰¤ Rating < 7.5: Average

      c) Rating â‰¥ 7.5: Excellent

```python
def categorize_rating(r):
  if r < 6:
    return "Poor"
  elif r < 7.5:
    return "Average"
  else:
    return "Excellent"

df["Rating_Category"] = df["Rating"].apply(categorize_rating)

df[["Movie_Title", "Rating", "Rating_Category"]]
```

```python
df["Rating_Summary"]=df["Rating"].apply(lambda x: 'Poor' if x < 6 else ('Average' if x < 7.5 else 'Excellent'))
df[['Movie_Title', 'Rating', 'Rating_Summary']]
```

###18) Create a custom function to score movies based on:

     i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score

     ii)Apply a function to dataset to assign the score

```python
def score_movie(row):
  return row["Rating"] * 10 + row["Revenue (Millions)"]

df["High_Score"] = df.apply(score_movie, axis=1)

df[["Movie_Title", "Rating", "Revenue (Millions)", "High_Score"]]
```

```python
df['High_Score'] = df.apply(lambda row: row['Rating'] * 10 + row['Revenue (Millions)'], axis=1)
df[['Movie_Title', 'Rating', 'Revenue (Millions)', 'High_Score']]
```
