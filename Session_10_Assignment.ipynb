```python
# Q 10

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
d = b**2 - 4*a*c
if d > 0:
  root1 = (-b + d**0.5) / (2*a)
  root2 = (-b - d**0.5) / (2*a)
  print("Two Real Roots:", root1, "and", root2)
elif d == 0:
  root = -b / (2*a)
  print("One Real Root:", root)
else:
  real = -b / (2*a)
  imag = (-d)**0.5 / (2*a)
  print("Complex Roots:")
  print(real, "+", imag, "i")
  print(real, "-", imag, "i")
```
